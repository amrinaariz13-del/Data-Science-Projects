```python
import pandas as pd

# Sample dataset
data = pd.DataFrame({
    'State': ['WA', 'WA', 'WA', 'NSW', 'NSW', 'NSW'],
    'Group': ['Kids', 'Men', 'Women', 'Kids', 'Men', 'Women'],
    'Sales': [20000, 25000, 15000, 18000, 22000, 17000]
})

# Group by State and calculate total sales
state_sales = data.groupby('State')['Sales'].sum().reset_index()
print("State-wise Total Sales:")
print(state_sales)

# Group by Group and calculate average sales
group_sales = data.groupby('Group')['Sales'].mean().reset_index()
print("\nGroup-wise Average Sales:")
print(group_sales)
```

```python

###### dataset:
data = pd.DataFrame({
    'State': ['WA', 'WA', 'WA', 'NSW', 'NSW', 'NSW'],
    'Group': ['Kids', 'Men', 'Women', 'Kids', 'Men', 'Women'],
    'Sales': [20000, 25000, 15000, 18000, 22000, 17000]
})

###### Group by State and calculate total sales
state_sales = data.groupby('State')['Sales'].sum().reset_index()
print("State-wise Total Sales:")
print(state_sales)

###### Group by Group and calculate average sales
group_sales = data.groupby('Group')['Sales'].mean().reset_index()
print("\nGroup-wise Average Sales:")
print(group_sales)
```

```python
import pandas as pd
import numpy as np

# Sample dataset
data = pd.DataFrame({
    'Date': ['2020-10-01', '2020-10-02', '2020-10-03', '2020-10-08', '2020-11-01', 
             '2020-11-05', '2020-12-10', '2020-12-15'],
    'State': ['WA', 'NSW', 'VIC', 'QLD', 'WA', 'NSW', 'VIC', 'QLD'],
    'Group': ['Kids', 'Men', 'Women', 'Seniors', 'Kids', 'Men', 'Women', 'Seniors'],
    'Sales': [20000, 25000, 10000, 37500, 7500, 15000, 7500, 27500],
    'Units': [8, 10, 5, 15, 3, 6, 3, 11]
})

# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Set 'Date' column as the index
data.set_index('Date', inplace=True)

# Weekly Report
weekly_report = data.resample('W').sum()

# Monthly Report
monthly_report = data.resample('M').sum()

# Quarterly Report
quarterly_report = data.resample('Q').sum()

# Display the reports
print("Weekly Report:")
print(weekly_report)

print("\nMonthly Report:")
print(monthly_report)

print("\nQuarterly Report:")
print(quarterly_report)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data as a DataFrame
data = {
    'Date': ['2020-10-01', '2020-10-02', '2020-10-03', '2020-10-04', '2020-10-05'],
    'Unit': [20.952381, 20.920635, 22.031746, 20.952381, 21.857143],
    'Sales': [20.952381, 20.920635, 22.031746, 20.952381, 21.857143]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Line chart
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Unit'], label='Units', marker='o', linestyle='--', color='blue')
plt.plot(df.index, df['Sales'], label='Sales', marker='x', linestyle='-', color='green')

# Add titles and labels
plt.title('Line Chart: Units and Sales Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Show plot
plt.show()
```
